```python
import pandas as pd
import re

# JSON 파일 로드
df = pd.read_json("../data/tokenized_data_제거3.json")

# 1️⃣ 영어가 포함된 title 필터링
df_with_english_title = df[df["title"].apply(lambda x: bool(re.search(r'[a-zA-Z]', str(x))))]

# 2️⃣ 필터링된 title 중에서 filtered_content에도 영어가 포함된 행만 선택
df_final = df_with_english_title[df_with_english_title["filtered_content"].apply(lambda x: bool(re.search(r'[a-zA-Z]', str(x))))]

# 3️⃣ 최종 기사 제목 출력
print(df_final["title"])
```
