```python
import json
import pandas as pd
```

```python
# 📌 3. JSON 데이터 로드
file_path = "../data/news_with_tokenized_keywords2.json"
with open(file_path, "r", encoding="utf-8") as f:
    data = json.load(f)

# 📌 4. 데이터프레임 변환
df = pd.DataFrame(data)
```

```python
df
```

```python


# 📌 필요 없는 컬럼 리스트
columns_to_drop = ["sentences", "filtered_sentences", "num_filtered_sentences", "num_sentences", "ratio", "keywords_type","trigrams_comb","trigrams_sw", "trigrams_perm", "bigrams_comb","bigrams_sw","bigrams_perm","bigrams","trigrams","keywords"]

# 📌 해당 컬럼 삭제 (존재하지 않아도 에러 발생하지 않도록 `errors="ignore"` 설정)
df.drop(columns=columns_to_drop, inplace=True, errors="ignore")
```

```python
df
```

```python
output_file = "../data/news.json"

# 📌 2. JSON 덤프를 위한 데이터 변환
json_data = df.to_dict(orient="records")  # DataFrame → JSON 리스트 변환

# 📌 3. Pretty Print JSON 파일 저장
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(json_data, f, indent=4, ensure_ascii=False)
```
