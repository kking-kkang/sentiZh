```python
import pandas as pd
import re

df = pd.read_json("../data/tokenized_data_제거3.json")
```

```python
# 검색할 키워드 지정
search_keyword = "台湾"

# 특정 키워드가 포함된 데이터 필터링
filtered_df = df[df["filtered_content"].str.contains(search_keyword, na=False)][["title", "quarter"]].sort_values(by="quarter")
```

```python
filtered_df
```

```python
# 특정 키워드가 등장한 분기만 필터링하여 개수 계산 및 정렬
quarter_counts = filtered_df["quarter"].value_counts().reset_index()
quarter_counts.columns = ["quarter", "키워드 등장 문서수"]
quarter_counts = quarter_counts.sort_values(by="quarter")

quarter_counts
```

```python
# 특정 키워드가 등장한 분기만 필터링하여 개수 계산 및 정렬
quarter_counts = filtered_df["quarter"].value_counts().reset_index()
quarter_counts.columns = ["quarter", "keyword_count"]

# 해당 분기의 전체 문서 개수 계산
total_counts = df["quarter"].value_counts().reset_index()
total_counts.columns = ["quarter", "total_count"]

# 두 개의 개수를 병합하여 해당 키워드가 등장한 분기만 유지
merged_counts = pd.merge(quarter_counts, total_counts, on="quarter", how="left").sort_values(by="quarter")

merged_counts
```
